```python
import pandas as pd 
import numpy as np


mr = pd.read_excel("..\Data\Tabular\All_Raagams.xlsx")

# mr = mr[mr['Melakarta']==1]
mr = mr[~mr['AAROHANAM'].isnull()]
mr = mr[mr['Raga Name'].apply(lambda x: "hindustani" not in str(x).lower())]
mr = mr.drop_duplicates(subset = ['AAROHANAM','AVAROHANAM'])


mr['path'] = mr['AAROHANAM'].apply(lambda x: str(x).replace("\xa0","-").split(" (Any")[0])
mr['avro_path'] = mr['AVAROHANAM'].apply(lambda x: str(x).replace("\xa0","-").split(" (Any")[0])
mr['pali'] = mr.apply(lambda x: x['path'].split("-") == x['avro_path'].split("-")[::-1] ,axis=1)
mr['value'] = 1
mr['raaga'] = mr['Raga Name']
mr['Raga_Rank'] = mr['Melakarta'].fillna(0).cumsum()
mr.to_csv("..\Data\Tabular\Raagams.csv", index = False)
```
