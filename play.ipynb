```python
import numpy as np
import pandas as pd
```

```python
mela_df = pd.read_excel("Melakarta Raagams.xlsx")
```

```python
mela_df.columns = ["_".join(i.strip().split()) for i in mela_df.columns]
```

```python
mela_df
```

```python
for n in ['AROHANAM','AVAROHANA']:
    mela_df[n]  = mela_df[n].apply(lambda x: x.replace("\n","").split())
```

```python
mela_df
```

```python
mela_df['SAME_AD'] = mela_df.apply(lambda x: x['AROHANAM']==x['AVAROHANA'][::-1], axis=1)
mela_df['STARTS_WITH_S'] = mela_df['AROHANAM'].apply(lambda x: x[0]=='S')
mela_df['MID1_WITH_S'] = mela_df['AROHANAM'].apply(lambda x: x[-1]=='S')
mela_df['MID2_WITH_S'] = mela_df['AVAROHANA'].apply(lambda x: x[0]=='S')
mela_df['ENDS_WITH_S'] = mela_df['AVAROHANA'].apply(lambda x: x[-1]=='S')
mela_df[mela_df['ENDS_WITH_S']==False]
```

```python
sargam = list('SRGMPDN')
sargam
```

```python
uniq_keys = sorted(set([i for x in mela_df['AROHANAM'].tolist() for i in x ]), key = lambda x: sargam.index(x[0]) + int(x[1])/10 if len(x)==2 else sargam.index(x[0])) + ['S']
uniq_keys
```

```python
for i in uniq_keys:
    mela_df[i] = mela_df['AROHANAM'].apply(lambda x: 1 if i in x else 0)
```

```python
display(mela_df[['MELA_NAME'] + uniq_keys])
```
